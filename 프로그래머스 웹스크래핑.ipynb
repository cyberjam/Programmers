# 전체 페이지 크롤링 및 웹스크래핑

```python
import urllib.request as REQ
import urllib.parse
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

# 한굴깨짐 방지
import os
os.environ['NLS_LANG'] = '.AL32UTF8'

urlpage_num = [str(i) for i in range(1, 11)]
urlpage = '/learn/challenges?tab=all_challenges/&page='
url1 = 'https://programmers.co.kr'

title = []
level_type = []
Link = []
level = []
for i in urlpage_num:
    url = url1 + urlpage + i

    response = REQ.urlopen(url)
    soup = BeautifulSoup(response, 'html.parser' )

    for div in soup.find_all('div', {'class':'col-item'}):
        t = div.find('h4', {'class':'title'}).getText().strip()
        l = div.find('h6',{'class':'level'}).getText().strip()
        lev = div.find('div')
        lev1 = str(list(div)[1])[27:34]
        L = div.find('div',{'data-original-title':'python3'})
        if L != None:
            b = str(L)[110:158]
            link = url + b
            Link.append(link)
        else:
            Link.append(L)
#     L = url + L
        title.append(t)
        level_type.append(l)
        level.append(lev1)
#     Link.append(L)
print(title,'\n\n', level_type, '\n\n', Link)

# header = ['TITLE', 'Level_Type', 'Link']
cosmatic = pd.DataFrame({'title':title,
                         'level_type':level_type,
                          'Level':level,
                          'Link':Link})
# , 'Link':rink
kk = cosmatic.dropna()
# kk.to_csv('C:/Users/CPB06GameN/Desktop/프로그래머스 문제.csv')
kk
kk.reset_index

# kk.to_csv('C:/Users/CPB06GameN/Desktop/프로그래머스 문제.csv', encoding='utf-8')
kk = kk.sort_values(['level_type','Level'], ascending=[False,True])
kk
```
